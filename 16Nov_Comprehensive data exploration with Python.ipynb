================================================== Image of Skewness==========================================

Then we can calculate Deviate from the normal distribution, Have appreciable positive skewness, Show peakedness.

You can Simply know what is skewness anda kurtosis :
1. Skewness is a measurement of the distortion of symmetrical distribution or asymmetry in a data set. Skewness is demonstrated on a bell curve when data points are not distributed symmetrically to the left and right sides of the median on a bell curve. If the bell curve is shifted to the left or the right, it is said to be skewed.
2. kurtosis is a statistical measure that is used to describe distribution. Whereas skewness differentiates extreme values in one versus the other tail, kurtosis measures extreme values in either tail. Distributions with large kurtosis exhibit tail data exceeding the tails of the normal distribution (e.g., five or more standard deviations from the mean). Distributions with low kurtosis exhibit tail data that are generally less extreme than the tails of the normal distribution.


================================================== Image of Kurtosis==========================================

```python
print('skewness : %f' %df_train['SalePrice'].skew())
print('kurtosis : %f' %df_train['SalePrice'].kurt())
```

=================================================Scatter Plot==============================================================

```python
#Using Variable SalePrice with GrLivArea
print(df_train['GrLivArea'].describe())
```

```python
#Scatter 1
#Version 1 : Making the Scatter Plot From 2 Column, x = GrLivArea and y = 'SalePrice'
var = 'GrLivArea'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis = 1)
data.plot.scatter(x = var , y = 'SalePrice', ylim = 0.008)
```

```python
#Version 2 : Making the Scatter Plot From 2 Column, x = GrLivArea and y = 'SalePrice'
data = pd.concat([df_train['SalePrice'], df_train['GrLivArea']], axis = 1)
data.plot.scatter(x = 'GrLivArea', y = 'SalePrice', ylim = 0.008)
```

```python
#Scatter 2 : Making the Scatter Plot From 2 Column, x = TotalBsmtSF and y = 'SalePrice'
data = pd.concat([df_train['TotalBsmtSF'], df_train['SalePrice']], axis = 1)
data.plot.scatter(x = 'TotalBsmtSF', y = 'SalePrice', ylim = (0.008))
```

=================================================Box Plot==============================================================

```python
#Using Column Variable OverallQual and SalePrice
data = pd.concat([df_train['OverallQual'], df_train['SalePrice']], axis = 1)
f , ax = plt.subplots(figsize = (8,6)) #Make an outer box
ax.set_title('Scatter Plot OverallQual and SalePrice')

fig = sns.boxplot(x = 'OverallQual', y = 'SalePrice', data = data) #Make content of Plot

fig.axis(ymin = 0, ymax = 800000)
```

```python
#Using Column Variable YearBuilt and SalePrice
data = pd.concat([df_train['SalePrice'], df_train['YearBuilt']], axis = 1)
f, ax = plt.subplots(figsize=(18,6))
ax.set_title('Box Plot of YearBuilt and SalePrice')
fig = sns.boxplot(x = 'YearBuilt', y = 'SalePrice', data = data) #Make content of Plot
fig.axis(ymin = 0, ymax = 900000)
plt.xticks(rotation=90)
```
